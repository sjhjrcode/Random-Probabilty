```python
import numpy as np
import matplotlib.pyplot as plt

def PMF_N(n, alpha=1):
    total_probability = 0
    for m in range(n, 100):  # Using 100 as an upper bound for summation, which should be sufficient for convergence
        total_probability += (alpha**m * np.exp(-alpha) / np.math.factorial(m)) * (1/(m+1))
    return total_probability

# Plotting PMF of N for alpha=1
values = range(20)  # plotting for n values from 0 to 19
probabilities = [PMF_N(n) for n in values]

plt.bar(values, probabilities, color='blue', alpha=0.7)
plt.xlabel('n')
plt.ylabel('P_N(n)')
plt.title('Marginal PMF of N for Î±=1')
plt.show()
```
